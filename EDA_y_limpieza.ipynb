### ¿cómo se correlacionan las variables del dataset?

```python
def find_high_correlation(corr_matrix, threshold=0.8):
    """
    Encuentra las pares de variables con correlación mayor a un umbral dado
    en una matriz de correlación dada.

    Args:
        corr_matrix (pd.DataFrame): Matriz de correlación
        threshold (float): Umbral para considerar una correlación alta.
            Por defecto es 0.8.

    Returns:
        List[tuple]: Lista de pares de variables con correlación mayor a
            `threshold`.
    """
    # Encontrar los índices de los valores absolutos más grandes
    idx = corr_matrix.abs().unstack().nlargest(len(corr_matrix)**2 - len(corr_matrix)).index
    # Filtrar los valores que superan el umbral de correlación
    high_corr = []
    for col1, col2 in idx:
        if col1 != col2 and corr_matrix.loc[col1, col2] > threshold:
            if (col2, col1, corr_matrix.loc[col2, col1]) not in high_corr:
                high_corr.append((col1, col2, corr_matrix.loc[col1, col2]))
    return high_corr
```

```python
corr_matrix = df_stats.corr()
high_corr = find_high_correlation(corr_matrix, threshold=0.8)
```
